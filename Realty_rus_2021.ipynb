```python
import pandas as pd
import plotly.express as px

import plotly.io as pio
pio.renderers.default = "browser"
```

```python
# Здесь нужно подставить правильный путь к файлу с датасетом
df = pd.read_csv('datasets/realty_2021_rus_data.csv.zip', sep=';')
```

```python
df
```

```python
df['month'] = df['date'].str[:-3]
```

```python
df_agg = (df.groupby(['month', 'rooms'])
    .agg({'price': 'mean', 
          'date': 'count', 
          'area': 'mean'})
    .reset_index()
    .rename(columns={'date':'number_sales'})
)
```

```python
df_agg['price_millions'] = (df_agg['price'] / 1000000).round(2)
df_agg['rooms_category'] = df_agg['rooms'].apply(lambda x: 
                        'Студия' if x==-1 else str(x)+'-комнатная')
```

```python
df_agg.head(5)
```

```python
fig = px.bar(df_agg, 
             x='rooms_category', 
             y='number_sales',
             color='price_millions',
             range_color=[1, 15],
             labels={'price_millions':'Price, m.rub.', 
                     'number_sales':'Number of sales'},
             animation_frame="month"
            )
fig.show()
```
